```python
unsta = df.groupby(['company_size', 'experience_level'])['salary_in_usd'].mean().unstack()
unsta
```

```python
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.countplot(df['work_year'],palette='Set2')
```

```python
#salary distribution variation across years
plt.figure(figsize=(10,7))
sns.set_style('darkgrid')
sns.histplot(x = 'salary_in_usd',hue='work_year',data=df,palette='Set2',kde=True)
```

# We can clearly see that over the years the salary and density of high salaried employees is increasing
# The difference can also be due to non-availabilty of data for 2021 and 2020

```python
salary_per_job=df.groupby('job_title')['salary_in_usd','job_title','company_size'].max().nlargest(10,columns='salary_in_usd')
salary_per_job
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y='job_title',x='salary_in_usd',ci=110,data=salary_per_job)
plt.yticks(rotation=45)
plt.show()
```
